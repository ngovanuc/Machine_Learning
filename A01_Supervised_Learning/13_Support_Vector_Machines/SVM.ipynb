## üßÆ III. C√¥ng th·ª©c to√°n h·ªçc (Classification)

M·ª•c ti√™u l√† t√¨m $\mathbf{w}, b$ sao cho:

$$
y_i(\mathbf{w}^\top \mathbf{x}_i + b) \geq 1
$$

‚Üí T·ªëi ∆∞u ho√° h√†m m·∫•t m√°t:

$$
\min_{\mathbf{w}, b} \frac{1}{2} \|\mathbf{w}\|^2
$$

N·∫øu kh√¥ng ph√¢n chia ƒë∆∞·ª£c ho√†n h·∫£o ‚Üí th√™m **Soft Margin**:

$$
\min_{\mathbf{w}, b, \xi} \frac{1}{2} \|\mathbf{w}\|^2 + C \sum_{i=1}^{n} \xi_i
$$

* $\xi_i$: sai s·ªë cho ph√©p
* $C$: tham s·ªë ƒëi·ªÅu ch·ªânh ƒë·ªô "c·ª©ng m·ªÅm" c·ªßa margin


## üåÄ IV. Kernel Trick ‚Äì Bi·∫øn h√¨nh kh√¥ng gian

Khi d·ªØ li·ªáu kh√¥ng tuy·∫øn t√≠nh, SVM d√πng **kernel trick** ƒë·ªÉ:

* Bi·∫øn d·ªØ li·ªáu sang kh√¥ng gian cao h∆°n ‚Üí **c√≥ th·ªÉ tuy·∫øn t√≠nh trong kh√¥ng gian m·ªõi**

| Kernel             | C√¥ng th·ª©c                     | Khi n√†o d√πng          |
| ------------------ | ----------------------------- | --------------------- |
| **Linear**         | $x^\top x'$                   | D·ªØ li·ªáu tuy·∫øn t√≠nh    |
| **Polynomial**     | $(x^\top x' + c)^d$           | D·ªØ li·ªáu c√≥ ƒë∆∞·ªùng cong |
| **RBF (Gaussian)** | $\exp(-\gamma \|x - x'\|^2)$  | Ph·ªï bi·∫øn nh·∫•t         |
| **Sigmoid**        | $\tanh(\alpha x^\top x' + c)$ | √çt d√πng h∆°n           |

## üß™ VI. Code v√≠ d·ª• v·ªõi `scikit-learn`

```python
from sklearn.svm import SVC
from sklearn.datasets import load_iris
from sklearn.metrics import classification_report

X, y = load_iris(return_X_y=True)
model = SVC(kernel='rbf', C=1.0, gamma='scale')  # ho·∫∑c kernel='linear'
model.fit(X, y)

y_pred = model.predict(X)
print(classification_report(y, y_pred))
```
