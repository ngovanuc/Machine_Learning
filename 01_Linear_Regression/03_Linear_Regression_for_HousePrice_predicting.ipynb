```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

## 1. Load dataset

```python
data_path = "E:/LEARNING/Machine_learning/Data/HousePrice/data.csv"
df = pd.read_csv(data_path)
df.head()
```

## 2. Preprocessing dataset

```python
# Trong bài này, chỉ lấy những thuộc tính như price (output), và các thuộc tính cho đầu vào (input) như là: bedrooms, bathrooms, view
columns = df.columns
print(columns)
```

```python
drop_columns = ['date', 'sqft_living', 'sqft_lot', 'floors', 'waterfront', 'condition', 'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'street', 'city', 'statezip', 'country']
df = df.drop(columns=drop_columns, axis=1)
df.head()
```

```python
# Loại bỏ giá trị Nan
df = df.dropna()
```

```python
x = df.iloc[:, 1:]
x.head()
```

```python
y = df.iloc[:, 0:1]
y.head()
```

```python
# Split dataset into training and testing
x_train, x_test, y_train, y_test = train_test_split(x, y, shuffle=True, test_size=0.2)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

## 3. Build and training model


```python
model = LinearRegression()
model.fit(x_train, y_train)
```

## 4. Predict

```python
predicted = model.predict(x_test)
```

```python
print(predicted)
```

```python
print(y_test)
```

## 5. Evaluation

Sai số tuyệt đối trên toàn bộ dữ liệu

```python
print(mean_absolute_error(y_test, predicted))
```

Bình phương sai số trên toàn bộ dữ liệu

```python
mean_squared_error(y_test, predicted)
```

Hệ số xác định r^2
- r2 nằm trong khoảng 0 đến 1
- càng gần 1 thì mô hình tốt, và ngược lại

```python
r2_score(y_test, predicted)
```
