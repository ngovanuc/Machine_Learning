```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

## 1. Tạo hoặc thu thập dữ liệu

```python
num_samples = 100
x = np.random.rand(num_samples) * 10

m = 2.5
c = 5
noise = np.random.rand(num_samples) * 2

y = m * x + c + noise
plt.scatter(x, y, color='blue', label='Dữ liệu ngẫu nhiên')
plt.plot(x, m * x + c, color='red', label='Đường y = mx + c')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

## 2. Tiền xử lý dữ liệu

```python
x.shape
```

```python
x = x.reshape((-1, 1))
x.shape
```

```python
y.shape
```

## 3. Xây dựng model và huấn luyện

```python
model = LinearRegression()
model.fit(x, y)
```

## 4. Predict hoặc đánh giá hiệu suất model

```python
# Dự đoán cho dữ liệu mới (ví dụ: dự đoán cho x = 8.5)
predicted = model.predict([[8.5]])
print(f"Predicted = {predicted}")

# Hãy so sánh kết quả predicted với hình vẽ để ước lượng độ chính xác
```
