> PCA l√† ‚Äúdao g·ªçt d·ªØ li·ªáu‚Äù gi√∫p **gi·∫£m chi·ªÅu**, **tƒÉng tr·ª±c quan h√≥a**, v√† ƒë√¥i khi gi√∫p **tƒÉng t·ªëc m√¥ h√¨nh**.


## üß† I. PCA l√† g√¨?

**Principal Component Analysis (PCA)** l√† m·ªôt k·ªπ thu·∫≠t **gi·∫£m chi·ªÅu kh√¥ng gi√°m s√°t**, bi·∫øn ƒë·ªïi d·ªØ li·ªáu t·ª´ kh√¥ng gian ban ƒë·∫ßu th√†nh **t·∫≠p t·ªça ƒë·ªô m·ªõi (c√°c tr·ª•c ch√≠nh)** sao cho:

* **Tr·ª•c ƒë·∫ßu ti√™n (PC1)** ch·ª©a **nhi·ªÅu ph∆∞∆°ng sai nh·∫•t**
* **Tr·ª•c th·ª© hai (PC2)** ch·ª©a ph∆∞∆°ng sai **l·ªõn th·ª© hai**, v√† **vu√¥ng g√≥c** v·ªõi tr·ª•c ƒë·∫ßu ti√™n
* ‚Ä¶

T√≥m l·∫°i:

> PCA t√¨m ra **t·∫≠p tr·ª•c t·ªça ƒë·ªô m·ªõi** m√† tr√™n ƒë√≥ d·ªØ li·ªáu **tr·∫£i ra nhi·ªÅu nh·∫•t**.


## üßÆ II. To√°n h·ªçc c·ªßa PCA

Gi·∫£ s·ª≠ ta c√≥ d·ªØ li·ªáu $X \in \mathbb{R}^{n \times d}$ (n samples, d features):

1. **Chu·∫©n h√≥a d·ªØ li·ªáu**: tr·ª´ trung b√¨nh t·ª´ng chi·ªÅu ‚Üí zero mean

2. **T√≠nh ma tr·∫≠n hi·ªáp ph∆∞∆°ng sai (covariance)**:

   $$
   \Sigma = \frac{1}{n} X^T X
   $$

3. **Ph√¢n t√≠ch gi√° tr·ªã ri√™ng (eigen decomposition)** ho·∫∑c **SVD**:

   * T√¨m eigenvectors (v): c√°c tr·ª•c ch√≠nh (principal components)
   * T√¨m eigenvalues (Œª): m·ª©c ƒë·ªô ph∆∞∆°ng sai ·ª©ng v·ªõi t·ª´ng tr·ª•c

4. **Gi·ªØ k tr·ª•c ƒë·∫ßu ti√™n** c√≥ eigenvalue l·ªõn nh·∫•t ‚Üí gi·∫£m chi·ªÅu

5. **Bi·∫øn ƒë·ªïi d·ªØ li·ªáu**:

   $$
   Z = X W_k
   $$

   V·ªõi $W_k \in \mathbb{R}^{d \times k}$: ma tr·∫≠n c√°c vector ri√™ng t∆∞∆°ng ·ª©ng v·ªõi k gi√° tr·ªã ri√™ng l·ªõn nh·∫•t


## üìê III. M·ª•c ti√™u c·ªßa PCA

T·ªëi ƒëa h√≥a ph∆∞∆°ng sai d·ªØ li·ªáu sau chi·∫øu:

$$
\max_W \|XW\|^2 \quad \text{v·ªõi } W^T W = I
$$

Ho·∫∑c:

> T·ªëi thi·ªÉu h√≥a sai s·ªë ph·ª•c h·ªìi d·ªØ li·ªáu g·ªëc t·ª´ kh√¥ng gian gi·∫£m chi·ªÅu.


## üß™ IV. Code v√≠ d·ª• b·∫±ng `sklearn`

```python
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

# Load data
X, y = load_iris(return_X_y=True)

# Gi·∫£m c√≤n 2 chi·ªÅu
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# V·∫Ω
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='Set1')
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA on Iris Dataset")
plt.show()

# T·ª∑ l·ªá ph∆∞∆°ng sai gi·ªØ l·∫°i
print("Explained variance ratio:", pca.explained_variance_ratio_)
```
