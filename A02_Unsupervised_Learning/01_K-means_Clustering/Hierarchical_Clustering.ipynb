* D·ªØ li·ªáu ƒë∆∞·ª£c nh√≥m theo c·∫•p b·∫≠c
* C√≥ th·ªÉ tr·ª±c quan b·∫±ng **c√¢y ph√¢n c·∫•p (dendrogram)**

## üìè II. C√°ch ho·∫°t ƒë·ªông c·ªßa Agglomerative Clustering

1. B·∫Øt ƒë·∫ßu: m·ªói ƒëi·ªÉm d·ªØ li·ªáu l√† m·ªôt c·ª•m ri√™ng
2. L·∫∑p:

   * T√≠nh kho·∫£ng c√°ch gi·ªØa c√°c c·ª•m
   * G·ªôp 2 c·ª•m g·∫ßn nhau nh·∫•t l·∫°i
3. L·∫∑p ƒë·∫øn khi ch·ªâ c√≤n 1 c·ª•m (ho·∫∑c ƒë·∫°t s·ªë c·ª•m mong mu·ªën)


## üìê III. Kho·∫£ng c√°ch gi·ªØa c√°c c·ª•m

| Ph√©p ƒëo              | C√¥ng th·ª©c                             | Ghi ch√∫          |
| -------------------- | ------------------------------------- | ---------------- |
| **Single Linkage**   | $\min d(x_i, x_j)$                    | D·ªÖ t·∫°o chu·ªói d√†i |
| **Complete Linkage** | $\max d(x_i, x_j)$                    | C·ª•m ch·∫∑t         |
| **Average Linkage**  | Trung b√¨nh kho·∫£ng c√°ch                | D·ªÖ hi·ªÉu          |
| **Ward‚Äôs Method**    | T·ªëi thi·ªÉu h√≥a t·ªïng ph∆∞∆°ng sai n·ªôi c·ª•m | R·∫•t ph·ªï bi·∫øn     |


## üî¨ IV. Dendrogram ‚Äì C√¢y ph√¢n c·∫•p

L√† bi·ªÉu ƒë·ªì th·ªÉ hi·ªán c√°ch c√°c c·ª•m ƒë∆∞·ª£c g·ªôp l·∫°i qua t·ª´ng b∆∞·ªõc.

D·ª±a v√†o **chi·ªÅu cao c·ªßa nh√°nh**, ta c√≥ th·ªÉ **ch·∫∑t c√¢y** ƒë·ªÉ ch·ªçn s·ªë c·ª•m ph√π h·ª£p.


## üß™ V. V√≠ d·ª• code v·ªõi `scikit-learn` + `scipy`

```python
from sklearn.datasets import load_iris
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

# Load d·ªØ li·ªáu
X, _ = load_iris(return_X_y=True)

# T√≠nh li√™n k·∫øt (linkage matrix)
Z = linkage(X, method='ward')

# V·∫Ω dendrogram
plt.figure(figsize=(10, 5))
dendrogram(Z)
plt.title("üìä Dendrogram - Hierarchical Clustering")
plt.xlabel("Samples")
plt.ylabel("Distance")
plt.show()

# Th·ª±c hi·ªán ph√¢n c·ª•m
model = AgglomerativeClustering(n_clusters=3, linkage='ward')
labels = model.fit_predict(X)
```
